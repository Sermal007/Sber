```python
data = df
cleardata = data.dropna()
data = cleardata.T
data['Unique_count'] = data.nunique(axis = 1)
data['Zero_count'] = data.eq(0).sum(axis=1)
data['Zero_percentage'] = round(data.Zero_count / data.shape[1] * 100, 2)
data['Nan_count'] = data.isna().sum(axis=1)
data['Nan_percentage'] = round(data.Nan_count / data.shape[1] * 100, 2)
data['Zero_percentage_from_non_nans'] = round(data.Zero_count / (data.shape[1] - data.Nan_count) * 100, 2)
data.iloc[2:, -6:]
```
